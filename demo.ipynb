```python
import pandas as pd
import numpy as np
from google.cloud import bigquery
```

```python
client = bigquery.Client(project="a-cubed-dev")

job_config = bigquery.QueryJobConfig(
    allow_large_results=True
)
```

```python
results = client.query(query="SELECT * FROM a-cubed-dev.raw_data.nyc_dcp_pluto LIMIT 1000", job_config=job_config)
df = results.to_dataframe()
```
